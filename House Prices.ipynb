### Create dummy variables for categoricals

```python
train_data.columns
```

#### Train dataset:

```python
categoricals = ['MSZoning', 'LandContour', 'LotConfig',
                'LandSlope', 'Neighborhood', 'BldgType', 'HouseStyle',
                'RoofStyle',
                'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
                'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
                'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                'Heating', 'HeatingQC',
                'CentralAir', 'Electrical',
                'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
                'KitchenQual',
                'Fireplaces', 'GarageType', 'GarageFinish', 'GarageCars',
                'GarageQual', 'GarageCond', 'PavedDrive',
                'MoSold', 'SaleType', 'SaleCondition']
```

```python
#GET_DUMMIES
#Encoding the categorical features
categoricals_oh = []

for label in categoricals:
    categorical_oh = pd.get_dummies(train_data[label], prefix=label)
    categoricals_oh.append(categorical_oh)
```

```python
#Comnine the encoded columns

df_encoded = pd.concat([train_data] + categoricals_oh, axis=1)
df_encoded.head()
```

```python
#Drop the original categorical features
df_encoded = df_encoded.drop(categoricals, axis=1)
df_encoded.head()
```

#### Test Dataset:

```python
#GET_DUMMIES
#Encoding the categorical features
test_categoricals_oh = []

for label in categoricals:
    test_categorical_oh = pd.get_dummies(test_data[label], prefix=label)
    test_categoricals_oh.append(test_categorical_oh)
```

```python
#Comnine the encoded columns

test_df_encoded = pd.concat([test_data] + test_categoricals_oh, axis=1)
test_df_encoded.head()
```

```python
#Drop the original categorical features
test_df_encoded = test_df_encoded.drop(categoricals, axis=1)
test_df_encoded.head()
```

#### Drop unique values

```python

not_in_test = [i for i in df_encoded.columns if i not in test_df_encoded.columns]
not_in_train = [i for i in test_df_encoded.columns if i not in df_encoded.columns]


y = df_encoded["SalePrice"]
df_encoded = df_encoded.drop(not_in_test, axis = 1)
test_df_encoded = test_df_encoded.drop(not_in_train, axis = 1)
df_encoded["SalePrice"] = y
```

### Building our Machine Learning Models

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_val)
```

**Desicion Tree**

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred = dt.predict(X_val)
```

```python
from sklearn import tree
dtr = tree.DecisionTreeRegressor()

dtr.fit(X_train, y_train)
y_pred = dtr.predict(X_val)
```
